# Problem 3

```python
import matplotlib.pyplot as plt

# distance formula in a function
def distance(x1,x2,y1,y2): 
    return ((x1 - x2)**2 + (y1 - y2)**2)**0.5
 
# x axis values
x = [2,-5]
# corresponding y axis values
y = [-3,9]
 
# plotting the points
plt.plot(x, y, color='green', linestyle='dashed', linewidth = 3,
         marker='o', markerfacecolor='blue', markersize=12)
 
# setting x and y axis range
plt.xlim(-10,10)
plt.ylim(-15,15)
 
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
 
# point a
x1 = 2
y1 = -3
# point b
x2 = -5
y2 = 9

# distance b/w a and b
d=distance(x1,x2,y1,y2)
 
# giving a title to my graph
# plt.title("Distance between points ({}, {}) and ({}, {}) is {}".format(x1,y1,x2,y2,distance))
plt.title("Distance 2 between points\n",fontweight="bold", fontsize=25, color='k')
plt.annotate("Distance between points \n({}, {}) and ({}, {}) is \n{}".format(x1,y1,x2,y2,d), xy=(-1.5, 3),  xycoords='data',
            xytext=(1,0.90), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='center',
            )
# function to show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(-5,5,100)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.legend(loc='upper right')
x = [2,2]
y = [-1,-2]
 
# plotting the points
X = np.array([1,3])
Y = np.array([1,4])
# Plotting point using scatter method
plt.scatter(X,Y,marker="x",c="blue")
plt.scatter(x,y,marker="o",c="red")

supp_vector_x = [1,2]
supp_vector_y = [1,-1]

Slope=(supp_vector_y[1]-supp_vector_y[0])/(supp_vector_x[1]-supp_vector_x[0])
# print(Slope)
perpindicular_line_slope=-1/Slope
# print(perpindicular_line_slope)
def midpoint(x1, x2, y1, y2):
    return ((x1 + x2)/2, (y1 + y2)/2)


a,b=midpoint(supp_vector_x[0],supp_vector_x[1],supp_vector_y[0],supp_vector_y[1])
plt.scatter(a,b,marker="*",c="Red",s=500)

pt = (a,b)
# print(pt)
# Given the above, calculate y intercept "b" in y=mx+b
b = pt[1] - perpindicular_line_slope * pt[0]
# Now draw two points around the input point
pt1 = (pt[0] - 5, perpindicular_line_slope * (pt[0] - 5) + b)
pt2 = (pt[0] + 5, perpindicular_line_slope * (pt[0] + 5) + b)
# Draw two line segments around the input point
plt.plot((pt1[0], pt[0]), (pt1[1], pt[1]), marker = 'o')
plt.plot((pt[0], pt2[0]), (pt[1], pt2[1]), marker = 'o')

sptv1=(1,1)
bs1 = sptv1[1] - perpindicular_line_slope * sptv1[0]
# Now draw two points around the input point
pt1s1 = (sptv1[0] - 5, perpindicular_line_slope * (sptv1[0] - 5) + bs1)
pt2s2 = (sptv1[0] + 5, perpindicular_line_slope * (sptv1[0] + 5) + bs1)
# Draw two line segments around the input point
plt.plot((pt1s1[0], sptv1[0]), (pt1s1[1], sptv1[1]), marker = 'o')
plt.plot((sptv1[0], pt2s2[0]), (sptv1[1], pt2s2[1]), marker = 'o')

sptv2=(2,-1)
bs2 = sptv2[1] - perpindicular_line_slope * sptv2[0]
# Now draw two points around the input point
pt1s1 = (sptv2[0] - 5, perpindicular_line_slope * (sptv2[0] - 5) + bs2)
pt2s2 = (sptv2[0] + 5, perpindicular_line_slope * (sptv2[0] + 5) + bs2)
# Draw two line segments around the input point
plt.plot((pt1s1[0], sptv2[0]), (pt1s1[1], sptv2[1]), marker = 'o')
plt.plot((sptv2[0], pt2s2[0]), (sptv2[1], pt2s2[1]), marker = 'o')

d2=distance(1,2,1,-1)
print(d2)
plt.plot(supp_vector_x, supp_vector_y, color='green', linestyle='dashed', linewidth = 3,
         marker='o', markerfacecolor='blue', markersize=12)
plt.title("SVM - Support Vectors - Margin - Distance\n",fontweight="bold", fontsize=25, color='k')
plt.annotate("Margin Distance btw support vectors: \n(1, 1) and (2, -1) is {}".format(d2), xy=(1.5, 0),  xycoords='data',
            xytext=(1,1), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='center',
            )

plt.xlim(-5, 5)
plt.ylim(-5, 5)
fig.set_figwidth(10)
fig.set_figheight(10)
plt.show()
```
